```python
def comb(n, k):
    if k == n:
        return [[i for i in range(1, n + 1)]]
    elif k == 1:
        return [[i] for i in range(1, n + 1)]
    ans = []
    for each_list in comb(n - 1, k - 1):
        ans.append(each_list + [n])
    for each_list in comb(n - 1, k):
        ans.append(each_list)
    return ans
```

```python
T = comb(10,5)
T
```

```python
def C(n,m):
    if m > n :
        return 0
    elif m == 1:
        return n
    elif n == 1:
        return 1
    else:
        return C(n-1,m-1)+C(n-1,m)
```

```python
C(10,5)
```
