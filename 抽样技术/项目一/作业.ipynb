```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\giao\Desktop\作业\抽样技术\diamonds.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data0 = data.sample(5000,replace = False)[['carat','price']].values
data0
```

```python
betas = []
reg = LinearRegression()
for i in range(500):
    data2 = data.sample(5000,replace = False)[['carat','price']].values
    x = data2[:,0].reshape(-1,1)
    y = data2[:,1].reshape(-1,1)
    reg.fit(x,y)
    beta = reg.coef_[0]
    betas.append(beta)
print(betas)
```

```python
beta = np.array(betas)
beta
```

```python
beta = np.sort(beta)
beta
```

```python
mean = np.mean(beta)
mean
```

```python
var = np.var(beta)
var
```

```python
s = np.std(beta, ddof=1)
xbar = np.mean(beta)
```

```python
#数据可视化 画数据经验分布曲线图
nt, bins, patches = plt.hist(beta, bins=10, histtype='step', \
cumulative=True, density=True, color='darkcyan')                    
plt.title('bins = 10')
plt.savefig('经验函数分布图1.jpg', dpi=200)
plt.show()
```

```python
#数据可视化 画数据经验分布曲线图
nt, bins, patches = plt.hist(beta, bins=15, histtype='step', \
cumulative=True, density=True, color='darkcyan')#beta是数据,bins是分组数
plt.title('bins = 15')
```

```python
#正态分布函数曲线拟合
y = (1 / (np.sqrt(2 * np.pi) * s)) * np.exp(-0.5 * ((bins - xbar) ** 2 / s ** 2))
y = y.cumsum()
y = y / y[-1]
plt.plot(bins, y, 'tomato', linewidth = 1.5)
plt.title('Theoretical')
plt.savefig('经验函数分布图2.jpg', dpi=200)
plt.show()
```
